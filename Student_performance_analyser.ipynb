```python
import numpy as np
import pandas as pn

#Data of students
data = { "Name" : ["Lana del rey" , "Vivian" , "Kin Young" , "Dante Russo" , "Angela Nikolau" , "Beerkus"],
         "English" : [89 , 90 , 78 , 88 , 98 , 88 ],
         "Mathematics" : [57 , 89 , 56 , 78 , 79 , 84],
         "Geography" : [76 , 65 , 78 , 89 , 56 , 67]}

df = pn.DataFrame(data)
df.index = [1,2,3,4,5,6]
print(df)
```

```python
#Average Marks
df["Average Marks"] = (df["English"] + df["Mathematics"] + df["Geography"]) / 3
print(df)
```

```python
#Topper 
top = df.loc[df["Average Marks"].idxmax(),"Name"]
print("Topper marks is" , top)
low = df.loc[df["Average Marks"].idxmin(),"Name"]
print("Lowest score is ",low)
```

```python
conditions = [
    df["Average Marks"] >= 80,
    (df["Average Marks"] >= 60) & (df["Average Marks"] < 80),
    df["Average Marks"] < 60
]

grades = ["A", "B", "C"]

df["Grade"] = np.select(conditions, grades, default="Not Assigned")

print(df)
```

```python
#Statistics Performance
print("Mean is ", np.mean(df["Average Marks"]))

print("Median is ",np.median(df["Average Marks"]))

print("Standar Deviation",np.std(df["Average Marks"]))

print("Percentile is ",np.percentile(df["Average Marks"],60))
```

```python
#Sorting
sort = df.sort_values(by = "Average Marks")
sort.index = [1,2,3,4,5,6]
print(sort)
```
